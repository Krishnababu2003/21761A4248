```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt
os.environ['TF_CPP_MIN_LOG_LEVEL'] ='2'
# Create needed objects
sgd = tf.keras.optimizers.SGD(learning_rate=0.1, momentum=0.9)
var = tf.Variable(0.5)
cost = lambda: 2 + var ** 2
x=[]
y=[]
# Perform optimization
# for _ in range(10):
#   sgd.minimize(cost, var_list=[var])
#   x.append(var.numpy())
#   y.append(cost().numpy())

for _ in range(10):
    with tf.GradientTape() as tape:
        loss = cost()
    gradients = tape.gradient(loss, [var])
    sgd.apply_gradients(zip(gradients, [var]))  # Apply gradients using the optimizer
    x.append(var.numpy())
    y.append(cost().numpy())

# Extract results
print(x)
print(y)
plt.plot(x,marker='o',markerfacecolor='orange',color='black')
plt.title('Gradient Descent Function')
plt.xlabel('Iteration')
plt.ylabel('Value of the parameter')
plt.show()
plt.plot(y)
plt.show()
plt.plot(x)
plt.show()
```
